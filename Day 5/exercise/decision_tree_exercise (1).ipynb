Decision Tree Clasification

```python
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
inputs = df.drop('Pclass', axis='columns')
```

```python
target = df['Pclass']
```

```python
inputs['Age'] = inputs['Age'].fillna(inputs['Age'].median())
```

```python
from sklearn.preprocessing import LabelEncoder

le_sex = LabelEncoder()
```

```python
inputs['Sex_n'] = le_sex.fit_transform(inputs['Sex'])
```

```python
inputs_n = inputs[['Survived', 'Sex_n', 'Age', 'Fare']]
inputs_n
```

```python
target
```

```python
from sklearn import tree

model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target)
```

```python
model.score(inputs_n, target)
```

What Pclass is a Survived female, Age 38, Fare 71.28?

```python
model.predict([[1, 0, 38, 71.28]])
```

What Pclass is a Not-survived male, Age 22, Fare 7.25?

```python
model.predict([[0, 1, 22, 7.25]])
```

What Pclass is a Survived male, Age 35, Fare 512?

```python
model.predict([[1, 1, 35, 512]])
```

```python
print("Model Accuracy:", round(model.score(inputs_n, target) * 100, 2), "%")
```
