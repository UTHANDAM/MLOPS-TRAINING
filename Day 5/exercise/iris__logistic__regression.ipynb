```python
from sklearn.datasets import load_iris
%matplotlib inline
import matplotlib.pyplot as plt
iris=load_iris()
```

```python
dir(iris)
```

```python
print("Target names:", iris.target_names)
print("Feature names:", iris.feature_names)
print("Data shape:", iris.data.shape)
```

```python
import numpy as np

feature_names = iris.feature_names

for i in range(5):
    plt.figure(figsize=(6, 2))
    plt.bar(feature_names, iris.data[i], color='steelblue')
    plt.title(f"Sample {i} — Label: {iris.target_names[iris.target[i]]}")
    plt.ylabel("cm")
    plt.tight_layout()
    plt.show()
```

```python
iris.data[0]
```

Create and Train Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
predictions = model.predict(iris.data[0:5])
print("Predicted labels:", predictions)
print("Predicted names:", iris.target_names[predictions])
```

Confusion Matrix

```python
y_predicted = model.predict(X_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10, 7))
sn.heatmap(cm, annot=True, fmt='d',
           xticklabels=iris.target_names,
           yticklabels=iris.target_names)
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.title('Confusion Matrix — Iris Logistic Regression')
plt.tight_layout()
plt.show()
```
